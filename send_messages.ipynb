```python
import pandas as pd
contacts = pd.read_csv("nachrichten_contacts.csv")
```

```python
# Import JSON module
import json
messages = {}
# Define JSON string
for i in range(5):
    file = open(f"outputs/savedata{i}.json")
    text = file.read()
    # Convert JSON String to Python
    file_messages= json.loads(text)
    messages.update(file_messages)


# Print Dictionary
print(messages)

```

```python
texts_sorted = []
for i in range(len(contacts["Given Name"])):
    name = contacts["Given Name"][i]
    name = name.replace(" ", "")
    texts_sorted.append(messages[name])
```

```python
contacts["Text"] = texts_sorted
```

```python
contacts.head()
```

```python
contacts.to_csv("nachrichten_contacts.csv", index=False)
```

```python
import pywhatkit
from datetime import datetime
import pyautogui




for index, row in contacts.iterrows():
    time = datetime.now()
    message = row["Text"]
    phone_number = str(row["Phone 1 - Value"]).replace(" ","")
    if phone_number.startswith("0"):
        phone_number = ("+41" +phone_number[1:])
    pywhatkit.sendwhatmsg(phone_number, message,int(time.strftime("%H")),int(time.strftime("%M"))+1)
    print(index)
    

# Displaying a Success Message
print("WhatsApp message sent!")
```
