```python
needed = pd.read_csv("nachrichten_contacts.csv")
```

```python
names = list(needed["Given Name"])
```

```python
print(names)
```

```python
slices =list(range(0,len(names),32))
slices.append(144)
print (slices)
```

```python
import os
API_KEY = os.getenv("API_KEY")
import json
from openai import OpenAI
client = OpenAI(
    api_key=API_KEY
)

for i in range(len(slices)-1):
    response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "user", "content": str(names[slices[i]:slices[i+1]]) + \
            'generiere für jeden Name in der obigen Liste einen etwa 50 Wörter langen Weihnachtstext mit Emojis, in welchem du schöne Weihnachten und ein gutes Neues Jahr wünschst. Gib eine Dictionnary im Format {"Name1":"Text1","Name2":"Text2",...,"NameX","TextX"} zurück. Unterschreibe mit Marco. Jeder Text soll sich in der Strukt einzigartig sein.'}
    ]
    )
    messages =json.loads(response.choices[0].message.content)
    save_file = open(f"outputs/savedata{i}.json", "w")  
    json.dump(messages, save_file, indent = 6, ensure_ascii=False)  
    save_file.close()  
    
```
